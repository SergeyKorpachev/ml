```python
a = 0.25

def loss1(y_true, y_pred):
    return k.mean( k.square( (y_pred - y_true) / (y_true) ) )

def loss2(y_true, y_pred):
    return k.mean( k.square( (y_pred - y_true) ) / (y_true) )

def loss3(y_true, y_pred):
    return k.mean( k.square( (y_pred - y_true) ) / ( a * y_true) )

def metric1(y_true, y_pred):
    return k.mean( ( y_pred - y_true ) / ( y_true ) )
```

```python
model = Sequential()
#model.add(Dense(32, activation='tanh', input_dim=data1.shape[1]))
model.add(Dense(32, activation='tanh', input_dim=x_train.shape[1]))
#model.add(Dense(32, activation='tanh', input_dim=data.shape[1]))
model.add(Dropout(0.1))
#model.add(Dense(32, activation='tanh'))
#model.add(Dropout(0.2))
model.add(Dense(10, activation='tanh'))
model.add(Dropout(0.1))
model.add(Dense(1))

'''
a = Input(shape=x_train.shape[1])
b = Dense(8)(a)
b = Dropout(rate=0.5)(b)
b = Activation('tanh')(b)
b = Dense(4)(a)
b = Dropout(rate=0.5)(b)
b = Activation('tanh')(b)
c = Dense(1)(b)
model = Model(inputs = a, outputs = c)
'''

#model.compile(loss='mean_squared_error', metrics=[metric1], optimizer='adam')
model.compile(loss=loss3, metrics=[metric1], optimizer='adam')

model.summary()
```
